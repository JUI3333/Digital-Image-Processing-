```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:/Users/Jui Sonawane/Desktop/FDIP/Range_Rover.jpg")
plt.axis("off")
plt.imshow(img)
```

```python
img = cv2.imread("C:/Users/Jui Sonawane/Desktop/FDIP/Range_Rover.jpg", cv2.IMREAD_GRAYSCALE)

cv2.imwrite("sunset_gray.jpg", img) 
plt.imshow(img, cmap='gray')   
plt.title("Gray Image")
plt.axis("off")
plt.show()
```

```python
bright_up = cv2.convertScaleAbs(img, alpha=1, beta=50)

plt.imshow(bright_up, cmap='gray')
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```

```python
bright_down = cv2.convertScaleAbs(img, alpha=1, beta=-50)

plt.imshow(bright_down, cmap='gray')
plt.title("Brightness Reduced")
plt.axis("off")
plt.show()
```

```python
negative = 255 - img

plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis("off")
plt.show()
```

```python
# Convert to float before log
img_float = img.astype(np.float32)

c = 255 / np.log(1 + np.max(img_float))
log_trans = c * np.log(1 + img_float)

# Convert back to uint8
log_trans = np.array(log_trans, dtype=np.uint8)

plt.imshow(log_trans, cmap='gray')
plt.title("Log Transformation (Fixed)")
plt.axis("off")
plt.show()
```

```python
gamma = 0.5   # try also >1
gamma_corrected = np.array(255 * (img / 255) ** gamma, dtype=np.uint8)

plt.imshow(gamma_corrected, cmap='gray')
plt.title("Gamma (0.5) Transformation")
plt.axis("off")
plt.show()
```

```python
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Thresholding")
plt.axis("off")
plt.show()
```
