```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\Jui Sonawane\Desktop\FDIP\FDIP Assignment 3\Range Rover 7.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Read image in grayscale
img = cv2.imread("C:/Users/Jui Sonawane/Desktop/FDIP/FDIP Assignment 3/Range Rover 7.jpg", cv2.IMREAD_GRAYSCALE)

# Display original image
plt.figure(figsize=(4,4))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Compute histogram
plt.figure(figsize=(5,3))
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Find min and max intensity values
r_min, r_max = np.min(img), np.max(img)

# Apply contrast stretching
stretched = ((img - r_min) / (r_max - r_min)) * 255
stretched = stretched.astype(np.uint8)

# Display contrast stretched image
plt.figure(figsize=(4,4))
plt.imshow(stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(stretched.ravel(), bins=256, range=[0,256], color='black')
plt.title("Contrast Stretched Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Apply histogram equalization
equalized = cv2.equalizeHist(img)

# Display equalized image
plt.figure(figsize=(4,4))
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(equalized.ravel(), bins=256, range=[0,256], color='black')
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Display side by side comparison
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(stretched, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

plt.show()
```
